```python
import sqlite3 
import csv 
import pandas as pd
```

```python
def create_db(db_name,file_name,table_name):
    con = sqlite3.connect(db_name)
    cursor = con.cursor()

    passenger=pd.read_csv(file_name)

    columns_to_drop=['Airport Country Code','Country Name','Airport Continent','Continents','Pilot Name']
    passenger.drop(columns_to_drop,axis=1,inplace=True)
    passenger=passenger[:1001]
    passenger.to_sql(table_name,con,if_exists='replace',index=False)

    results=cursor.execute(f"select * from {table_name}").fetchall()
    con.commit()
    con.close()
    
    
    return(results)
```

```python
passenger=create_db("Airlines.db","AirlineDataset.csv","passenger")
```
